LIMIT_BAL,BILL_AMTx and PAY_AMTx are meeant to be in dollars, and should be converted to a numerical data type to reflect this. 

```python
# Convert the columns identified above to numeric

# Get list of numeric data types
billColumns = ["BILL_AMT{}".format(x) for x in range(1,7)]
payAmtColumns = ["PAY_AMT{}".format(x) for x in range(1,7)]

numericColumns = ['LIMIT_BAL','AGE']+billColumns+payAmtColumns

# Apply numeric data types
for i,col in enumerate(numericColumns):
    df[col] = pd.to_numeric(df[col])
```

### Part 2 : Apply Data Cleaning

#### ID

```python
# Remove ID Column, it is identical to index
df.drop('ID' , axis=1, inplace=True)
```

```python
df
```

```python
# Drop duplicates
df = df.drop_duplicates()
```

```python
df
```

#### SEX

```python
# Apply one-hot encoding style to the feature sex, generating new feature gender_male
df['gender_male'] = df['SEX'] == 1
```

```python
df['gender_male']
```

#### Education

```python
# remove missing/unknown values in education feature with category 4. Other
df['EDUCATION'] = df['EDUCATION'].replace([0,5,6], 4)
```

```python
df['EDUCATION'].value_counts().sort_index() 
```

```python
# Apply one hot encoding to Education
dfEncodedEducation = pd.get_dummies(df['EDUCATION'], columns=['EDUCATION'])
```

```python
# rename to more human readable name
dfEncodedEducation = dfEncodedEducation.rename(columns = {1:"education_graduate_school", 2:"education_university", 3:"education_high_school", 4: "education_others"})
```

```python
dfEncodedEducation
```

```python
#combine with origional dataset
df = df.merge(dfEncodedEducation[['education_graduate_school', 'education_university', 'education_high_school', 'education_others']], left_index=True, right_index=True)
```

```python
df
```

#### MARRIAGE

```python
# MARRIAGE
df['MARRIAGE'].value_counts().sort_index() 
```

```python
# Remove missing column
df['MARRIAGE'] = df['MARRIAGE'].replace([0], 3)
```

```python
df['MARRIAGE'].value_counts().sort_index() 
```

```python
dfMarriage = pd.get_dummies(df['MARRIAGE'], columns=df['MARRIAGE'])
```

```python
dfMarriage = dfMarriage.rename(columns = {1:'marriage_married', 2:'marriage_single', 3:'marriage_other'})
```

```python
dfMarriage
```

```python
df = df.merge(dfMarriage, left_index=True, right_index=True)
```

```python
df
```

#### Age

```python
bill_amts_normalised = df[['BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6']]/df["LIMIT_BAL"]
```
